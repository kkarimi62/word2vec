```python
import numpy as np
import networkx as nx
import matplotlib.pylab as plt
import json
import gensim.downloader as api
```

```python
def PlotGraph(G):
    subax1 = plt.subplot(121)
    nx.draw(G, with_labels=True, font_weight='bold')
    subax2 = plt.subplot(122)
    nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')

def Distance(pos,i,j):
    return (np.sum((pos[i]-pos[j])**2))#**0.5

def withinThreshold(G, thresh):
    for item in G.edges():
        dist = Distance(pos,item[0],item[1])
        assert dist <= thresh, '%s <= %s!'%(dist,thresh) 
```

```python
import random
n = 10
dim=2
thresh=1
w = {i: 0.5 for i in range(n)}
pos = np.random.random(size=n*dim).reshape((n,2))
G = nx.geographical_threshold_graph(n, 1.0/thresh,dim=dim,pos=pos, weight=w)

PlotGraph(G)
withinThreshold(G, thresh)
```
